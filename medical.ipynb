```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("medical_insurance.csv") 
data.head()
```

```python
null_data = data.isnull().sum() 
null_data = null_data.to_frame() 
null_data.columns=['NullValues'] 
null_data 
```

```python
invalid_val = (data[['age','children','bmi','charges']]<0).sum().to_frame() 
invalid_val.columns = ['InvalidValues'] 
invalid_val
```

```python
min_val = data[['age','children','bmi','charges']].min().to_frame() 
min_val.columns = ['Min'] 
min_val['Max'] = list(data[['age','children','bmi','charges']].max()) 
min_val
```

```python
data.describe()
```

```python
data['charges'].replace(data['charges'].values,list(map(int,data['charges'])),inplace = True) 
data['charges'] = data['charges'].astype('int64') 
```

```python
sw_data = data[data['region'] =='southwest']
sns.lmplot(x='age', y='bmi', data=data,line_kws={'color':'red'}) 

sw_data = data[data['region'] =='southeast']
sns.lmplot(x='age', y='bmi', data=data,line_kws={'color':'red'}) 

sw_data = data[data['region'] =='northwest']
sns.lmplot(x='age', y='bmi', data=data,line_kws={'color':'red'}) 

sw_data = data[data['region'] =='northeast']
sns.lmplot(x='age', y='bmi', data=data,line_kws={'color':'red'}) 
```
